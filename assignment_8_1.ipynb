```python
# 1. barplots
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = pd.read_csv("Titanic.csv")

# Create count of passengers by sex
sex_counts = titanic['Sex'].value_counts()

# Create subplots
fig, ax = plt.subplots(1, figsize=(2, 3))

ax.bar(sex_counts.index, sex_counts.values, alpha=0.6, width=0.5)
ax.set_title('Passenger Count by Sex')
ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Histogram
fig, ax = plt.subplots()

plt.hist(titanic['Age'], 10, alpha = 0.6)
```

```python
# Boxplots
fig, ax = plt.subplots(figsize = (8, 6))

sns.boxplot(data = titanic, x = 'Pclass', y = 'Age', width = 0.1, 
            boxprops = dict(alpha=0.6)) 

ax.set_xlabel("Cabin class")

plt.show()
```

```python
# Boxplots
fig, ax = plt.subplots(figsize = (8, 6))

sns.boxplot(data = titanic, x = 'Survived', y = 'Age', width = 0.1, 
            boxprops = dict(alpha=0.6)) 

ax.set_xlabel("Survival")

plt.show()
```

```python
fig, ax = plt.subplots()

plt.hist(titanic['Fare'], 10, alpha = 0.6)

plt.xlabel('Fare')
plt.ylabel('Count')

plt.show()
```

```python
# Family/ticket class
fig, ax = plt.subplots(figsize=(10, 6))

sns.violinplot(data = titanic, x = 'Pclass', y = 'SibSp', ax = ax, width = 0.3)

ax.set_xlabel("Cabin class")

plt.show() 
```

```python
#stacked barplots
grouped = titanic.groupby(['Pclass', 'Survived']).size().unstack()

# Create stacked barplot
grouped.plot(kind='bar', stacked=True, figsize=(8, 5), colormap='Blues')

# Labels and title
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# violin chart
fig, ax = plt.subplots(figsize=(10, 6))

sns.violinplot(data = titanic, x = 'Sex', y = 'Age', ax = ax, width = 0.3)

ax.set_xlabel("Gender")

plt.show() 
```
